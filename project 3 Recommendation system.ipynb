```python
users_df = pd.read_csv(r"C:\\Users\\shiva\\Downloads\\Users.csv")
ratings_df = pd.read_csv(r"C:\\Users\\shiva\\Downloads\\Ratings.csv")
books_df = pd.read_csv(r"C:\\Users\\shiva\\Downloads\\Books.csv")

# Merge the DataFrames based on common columns 
Combined_df = pd.merge(users_df, ratings_df, on='User-ID')
Combined_df = pd.merge(Combined_df, books_df, on='ISBN')
```

```python
Combined_df.to_csv("Combined_data.csv", index=False)

Combined_data = pd.read_csv("Combined_data.csv")
Combined_data.drop(Combined_data.columns[9:12], axis=1, inplace=True)

Combined_data.head()
```
